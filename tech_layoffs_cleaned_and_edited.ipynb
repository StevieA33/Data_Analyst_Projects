```python
# convert the date column into a datetime object
new_df['reported_date'] = pd.to_datetime(new_df['reported_date'])
```

```python
# extract the day, month, and year components
new_df['day'] = new_df['reported_date'].dt.day
new_df['month'] = new_df['reported_date'].dt.month
new_df['year'] = new_df['reported_date'].dt.year
```

```python
# Searches the table for any duplicates and removes them from the dataframe
new_df.drop_duplicates()
```
