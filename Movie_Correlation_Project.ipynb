```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'C:\Users\zygophythewhale267\Documents\Portfolio_Projects\movies.csv')
```

```python
df.head()
```

```python
# This loops through the data and sees if there is anything missing, by looping through the columns and seeing if anything had a null in it or nothing
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Lists the data types for the columns

print(df.dtypes)
```

```python
#Correctly identifies the year the movie was released, by taking the month from the released column and placing it in the newly created column called monthreleased
df['monthreleased'] = df['released'].astype('str').str[:4]
df
```

```python
# Orders the table by the gross column in descending order
df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
# Searches the table for any duplicates and removes them from the dataframe
df.drop_duplicates()
```

```python
# Testing to see if budget will have a high correlation with the gross column by building a scatterplot
plt.scatter(x = df['budget'], y = df['gross'])
plt.title("Budget vs Gross Earnings")
plt.xlabel("Gross Earnings")
plt.ylabel('Budget for Film')
plt.show()
```

```python
# To determine if budget and gross columns are correlated by using regplot, because in the above scatterplot it's a little hard to tell if they correlate
sns.regplot(x="gross", y="budget", data=df)
```

```python
# Shows how much correlation is between the numeric columns of the dataframe in a more visual way
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
# This shows the same information above in a more visual way using the the correlation method called Pearson
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Shows 
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
# Shows the columns that have a high correlation (> 0.5)
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)

```
